import Required libraries :

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

Q2 :

```python
#Read image
img = cv2.imread("drive/MyDrive/DIP_EXC2/Q2/Q2_img.png")

#Map input intensity level to output intensity level
def Intensity(pix, a, Ya, b, Yb, l, m, n):

    if (0 <= pix and pix <= a):
        return (b / a)* l* pix

        
    elif (a < pix and pix <= b):
        return ((Yb - Ya)/(b - a))* m*(pix - a) + Ya

    else:
        return ((255 - Yb)/(255 - b))* n*(pix - b) + Yb
  
#Define parameters:
a = 40
Ya = 20
b = 130
Yb = 190
l = 0.1
m = 1.2
n = 1

#Vectorize the function 
Intensity_vec = np.vectorize(Intensity)
  
#Apply contrast stretching
contrast_img = Intensity_vec(img, a, Ya, b, Yb, l, m, n)

#Show image
horizontal_concat = np.concatenate((img, contrast_img), axis=1)
cv2_imshow(horizontal_concat)

#Save image
cv2.imwrite("drive/MyDrive/DIP_EXC2/Q2/Q2_contrast.png", contrast_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
